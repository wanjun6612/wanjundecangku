# Maven使用入门

## 1. 1 编写POM

     1. POM定义了项目基本信息，用于描述项目如何构建，声明项目依赖等。

     2. xml头指定了此文件的版本和编码方式。project是所有pom.xml的根元素，它还声明了一些POM相关的命名空间及xsd元素，虽然这些属性不是必须的，但使用这些属性能让第三方工具（如IDE中的XML编辑器）帮助我们快速编辑POM。
     3. modleVersion指定当前POM模型的版本。通常是4.0.0.这段代码中最重要的是包含groupId、artifactId和version的三行。这三个元素定义了一个项目基本的坐标。在Maven的世界里。任何的jar、pom或war都是基于这些基本的坐标进行区分的。
     4. groupId定义项目属于哪个组，这个组往往和项目所在的组织或公司存在关联。
     *  articactId定义了当前Maven项目在组中唯一的ID。你可能会为不同的子项目模块分配artifacIc.
     * version指定当前版本。
     * name元素声名了一个对于用户更为友好的项目名称，虽然不是必须。
     * 没有任何实际的JAVA代码，我们就能够定义一个Maven项目的POM，这体现了Maven的一大优点，它能让项目对象模型最大程度地与实际代码相独立，我们可以称之为解耦或正交性。
     
## 1. 2 编写主代码
    
     1. 编写主代码：主代码会打包到最终构件(jar),而测试代码只运行测试时用到，不会被打包。默认情况下，Maven假设项目主代码位于src/main/java目录，我们遵循Maven的约定，创建此目录。创建java文件。
    
```
    
    package com.wanjun.maven.helloworld

     public class helloWorld{

	public String sayHello(){
		return "hello maven";
	}
    public static void main(String[] args) {
	    System.out.print(new HelloWorld().sayHello());	
	}
    }

```
     2. 项目主代码放入src/main/java/包目录结构（遵循约定），无需额外配置。它会自动搜索词目录。其次java类的包命名与之前的POM中的groupId和artifactId吻合。这样结构更清晰，更符合逻辑，也方便搜索构建或java类    
     * 在项目根目录下运行命令 mvn clean compile得到编译输出
