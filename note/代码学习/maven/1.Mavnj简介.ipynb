# Maven简介

    1.Maven是一跨平台的项目管理工具。主要服务于Java平台的项目构建、依赖管理和项目信息管理。

## 1.1何为构建
     
     1.编译、运行单元测试、生成文档（测试报告）、打包和部署等这就是构建。
     
## 1.2 Maven是优秀的构建工具

     1. Maven抽象了一个完整的构建生命周期模型，这个模型吸取了大量其他的构建脚本和构建工具的优点。

## 1.3 为什么需要Maven 
     
     1.Make也许是最早的构建工具，但不是跨平台的。
     2.Ant不是指蚂蚁，而是意指“另一个整洁的工具”，它最早用来构建著名的Tomcat。其诞生的动机就是作者受不了Makefile的语法格式。可以将Ant看成一个Java版本的Make,也正因为是JAVA，所以Ant是跨平台的。此外Ant使用XML定义构建脚本，相对于Makefile来说，这也更加友好。与Make类似，Ant有一个构建脚本build.xml.

     3.build.xml的基本结构也是目标（target）、依赖（depends），以及实现目标的任务。比如在上面的脚本中，jar目标用来创建应用程序jar文件，此目标依赖于compile目标，后者执行的任务是创建一个名为classes的文件夹，编译当前目录的java文件至classes目录。Ant有大量内置用java实现的任务，这保证了其跨平台的特质。和Make一样，Ant也是过程式的，开发者显式的指定每一个目标，以及完成该目标所需要执行的任务。针对每一个项目，开发者都需要重新编写这一过程。Maven是声名式的，项目构建过程和过程各个阶段所需要的工作都是由插件实现，且大部分插件都是现成的，开发者只需要声名项目的基本元素。Maven就执行内置的、完整的构建过程。消除了重复。
     Ant是没有依赖管理的，用户不得不手工管理依赖。后来才出现了Ivy管理依赖。而Maven用户内置了依赖管理，更有一个可能拥有全世界最多的java开源软件包的中央仓库，Maven用户无需进行任何配置就可以直接想用。
  
## 1.4 不重复发明轮子
   
   1.Ant打包无非就是创建目录，复制文件，编译源代码，使用一堆任务。所有项目如果都遵守一套目录结构ant的bulid文件就可以复用。但这些功能Maven已提供。

### 1.4.1 Maven与极限编程
 
   1. XP编程强调的拥抱变化。
   * 简单：Maven暴露了一组一致、简介的操作接口，帮助团队成员从原来高度自定义、复杂的构建系统中解脱出来。
   * 交流与反馈。与版本控制系统结合后，所有人都能执行最新的构建并快速得到反馈。
   * 测试驱动开发：测试时Maven生命周期最重要的组成部分之一，且支持流行的测试框架。
   * 十分钟构建。
   * 持续集成：CI强调项目以很短的周期（如15分钟）集成最新的代码。实际CI的前提是源码管理系统和构建系统。目前业界流行的CI服务器都能很好的和Maven进行集成。
   * 富有信息的工作区：使用Maven发布的项目报告站点，并配置你需要的项目报告，如测试覆盖率报告，都能帮你把信息推送到开发者眼前。
