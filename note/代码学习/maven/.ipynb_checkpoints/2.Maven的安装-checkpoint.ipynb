# Maven 的安装和配置

##  1.1在Windows上安装Maven

    1. 运行如下命令来检查Java安装：java-version
    
##  1.2下载Maven
    1. http://maven.apache.org/download.html
    * 设置环境变量，将Maven在那装配置到操作系统环境中
    * 桌面右键“我的电脑”->“属性”，新建一个变量名为M2_HOME，值为Maven的安装目录
    D:\apache-maven-3.5.4。接着在系统变量中找到一个名为Path的变量在变量值末尾加上%M2_HOME%\bin
    * 验证    echo %M2_HOME%              mvn -v
    * 升级maven。下载->环境变量重设定即可。

##  1.3 基于Unix的系统上安装Maven

     1. 先检查java环境变量
     *  下载压缩包后解压。虽然直接使用此目录配置环境变量之后就能使用Maven，但推荐做法是在安装目录平行      的创建一个符号链接方便日后升级。
     * $tar -xvzf apache -maven-3.5.4-bin.tar.gz
     * $ln -s apache-maven-3.5.4 apache-maven
     * ls-l
     * 接下来需要设置M2_HOME环境变量指向符号链接apache-maven,并将Maven安装目录下的bin文件添加到系统环境变量path中：
     * export M2_HOME=/home/juven/bin/apache-maven
     * export PATH=$ PATH:$M2_HOME/bin
     * 将这两行命令加入到系统的登录shell脚本中去，每次启动一个终端，这些配置就能自动执行。
     * 可以利用符号链接这一工具简化Maven升级，不用像在Windows那样更新环境变量。
     
## 1.4 安装目录分析
     
 
      1. bin:包含mvn运行脚本，这些脚本用来配置JAVA命令，准备好classpath和相关的Java系统属性，然后执行Java命令。mvn是shell脚本，mvn.bat是bat脚本。在命令行输入任何一条mvn命令时，实际就是在调用这些脚本。 还包含debug的脚本。
      * boot:只包含一个文件：plexus-classworlds-2.2.3.jar。这是一个类加载器框架，相对于默认的java类加载器，它提供了更丰富的语法以方便配置。
      * conf:目录包含一个重要文件settings.xml，直接修改此文件就能在机器上全局定制Maven行为。一般情况下偏向复制此文件至~/.m2/目录下（~表示用户目录），在用户范围定制Maven行为。
      * lib:包含maven运行时所需要的类库。maven本身是分模块开发的，因此能看到不同的jar包。lib目录就是真正的maven。
      * ~/. m2：mvn help:system. 此命令会打印出所有java系统属性和环境变量。运行这条命令的目的是让maven直行一个真正的任务。我们可以从命令行输出看到maven会下载maven-help-plugin，包括pom文件和jar文件下载到Maven本地仓库。
      * 默认情况在用户目录下可以发现.m2文件夹，防止了本地仓库repository。所有构件都被存到此仓库方便重用。

## 1.5 设置HTTP代理
 
     1. 运行ping.repol.maven.org检查网络是否畅通。不畅通再设定http代理让其访问外部仓库。

## 1.7 Maven安装实践
     
     1. 一些安装Maven过程中不是必须，但十分有用的实践。
     * 运行mvn命令实际上是执行java命令。那么运行java命令可用到的参数当然也可以在运行mvn命令使用。
     * 通常设置MAVEN_OPTS值为 -Xms128m -Xmx512m
     * 配置用户范围settings.xml：~/.m2/settings.xml影响的是个人。统一的配置使用M2_HOME/conf/settings.xml
     * 不要使用IDE内嵌的MAVEN，因版本较新，往往不稳定。
